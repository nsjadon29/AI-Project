```python
# import pandas library 
import pandas as pd 
  
# Reading all the data about the movie from the file
column_names = ['user_id', 'item_id', 'Rating'] 
  
path = '/Users/nikhilsinghjadon/Desktop/fffile.tsv'
  
df = pd.read_csv(path, sep='\t',names=column_names) 
  
# Check the head of the data 
df.head(150)

# Check out all the movies and their respective IDs 
movie_names = pd.read_csv('/Users/nikhilsinghjadon/Desktop/Movie_Titles.csv') 
movie_names.head(150)

data = pd.merge(df, movie_names, on='item_id') 
data.head(150) #List of all movies
```

```python
# Calculate mean rating of all movies 
data.groupby('Movie')['Rating'].mean().sort_values(ascending=False).head() 

# Calculating  'average rating' count values 

ratings = pd.DataFrame(data.groupby('Movie')['Rating'].mean())  
ratings['Score'] = pd.DataFrame(data.groupby('Movie')['Rating'].count()) 
ratings.head(5) 
```
